```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/Taj112/PDS_Project/refs/heads/main/term-test-1-result.csv')
df1
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/Taj112/PDS_Project/refs/heads/main/term-test-2-result.csv')
df2
```

```python
merged_df = df1.merge(df2, on=["Registration Number", "Names"], how="inner")
merged_df
```

```python
merged_df['Best_Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)
merged_df['Average_Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1).round(2)
merged_df
```

```python
merged_df.drop(['TT-1 Marks', 'TT-2 Marks'], axis=1, inplace=True)
merged_df
```

```python
df3= pd.read_csv('https://raw.githubusercontent.com/Taj112/PDS_Project/refs/heads/main/Attendance%26Final.csv')
final_df = merged_df.merge(df3, on=["Registration Number"], how="inner")
final_df['Final Marks'] = final_df['Term_final'] * 0.7 + final_df['Average_Marks'] + final_df['Attendance']
final_df
final_df.to_csv("final result.csv", index=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('final result.csv')

final_marks = data['Final Marks'].values.reshape(-1, 1)

K_values = range(2, 6)
for K in K_values:
    kmeans = KMeans(n_clusters=K, random_state=42)
    data[f'Cluster_K{K}'] = kmeans.fit_predict(final_marks)


    plt.figure(figsize=(8, 6))
    for cluster in range(K):
        plt.scatter(data[data[f'Cluster_K{K}'] == cluster]['Final Marks'],
                    np.zeros_like(data[data[f'Cluster_K{K}'] == cluster]['Final Marks']),
                    label=f'Cluster {cluster + 1}')
    plt.title(f'K-Means Clustering with K={K}')
    plt.xlabel('Final Marks')
    plt.yticks([])
    plt.legend()
    plt.show()
```
